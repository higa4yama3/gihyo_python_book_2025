# 低緯度オーロラが出現する 未来や過去を眺めよう! 現在から江戸・鎌倉・紀元前まで、オーロラ出現をデータ分析

## 米海洋大気局(NOAA)のオーロラ予測情報を取得して表示するコード例と実行例

```python
import json; from urllib.request import urlopen
from datetime import datetime; import numpy as np
import matplotlib.pyplot as plt; import cartopy.crs as ccrs
from matplotlib.colors import LinearSegmentedColormap
from cartopy.feature.nightshade import Nightshade

def aurora_forecast():
    url='https://services.swpc.noaa.gov/json/ovation_aurora_latest.json'
    response = urlopen(url)
    aurora = json.loads(response.read().decode('utf-8'))
    dt=datetime.strptime(aurora['Forecast Time'],'%Y-%m-%dT%H:%M:%SZ')
    aurora_data = np.array(aurora['coordinates'])
    img = np.reshape(aurora_data[:, 2], (181, 360), order='F')
    img_proj = ccrs.PlateCarree(); img_extent = (0, 359, -90, 90)
    return img, img_proj, img_extent, 'lower', dt

def aurora_cmap(): # オーロラ風のカラーマップを作る
    stops = {'red': [(0.00,0.1725,0.1725),(0.50,0.1725,0.1725),(1.00,0.8353,0.8353)],
    'green':[(0.00,0.9294,0.9294),(0.50,0.9294,0.9294),(1.00,0.8235,0.8235)],
    'blue':[(0.00,0.3843,0.3843),(0.50,0.3843,0.3843),(1.00,0.6549,0.6549)],
    'alpha': [(0.00,0.0,0.1),(0.50,1.0,1.0),(1.00,1.0,1.0)]}
    return LinearSegmentedColormap('aurora', stops)

fig = plt.figure(figsize=[10, 5])
ax1 = fig.add_subplot(1, 2, 1, projection=ccrs.Orthographic(0, 90))    # 北半球
ax2 = fig.add_subplot(1, 2, 2, projection=ccrs.Orthographic(180, -90)) # 南半球
img, crs, extent, origin, dt = aurora_forecast()
for ax in [ax1, ax2]:
    ax.coastlines(zorder=3); ax.stock_img() # デフォルト画像を表示
    ax.gridlines();ax.add_feature(Nightshade(dt))
    ax.imshow(img, vmin=0, vmax=100, transform=crs,
        extent=extent, origin=origin, zorder=2,cmap=aurora_cmap())
plt.show()
```

## Kp 指数を使い、 太陽活動の程度に応じた「オーロラ出現領域」を表示するコードと実行例

### （一括実行用の）記事掲載コード

```python
from geopy.distance import geodesic # 緯度経度からの距離算出用
from scipy.optimize import curve_fit
import cartopy.geodesic ascgeod;import cartopy.feature as cfeature

# 2点間の緯度経度から緯度相当値を算出
def coord2mlat(lat,lon,mlat,mlon): # 指定場所と北磁極の緯度・経度
    return 90-geodesic((lat,lon),(mlat,mlon)).kilometers/10000*90
def mlat2km(mlat): # 緯度相当値mlatから距離を算出
    return 10000.0/90*(90-mlat)
kps  = np.array([5,       5,      7,       7,      9])
lons = np.array([-122.30,-79.40,-104.98, -86.78,-97.03])
lats = np.array([47.45,   43.65,  39.76,  36.16, 32.90])
mlats=[]              # 指定場所と北磁極間の角度を格納する
mlat=81.3;mlon=-110.8 # 現在の北磁極は北緯81.3°西経110.8°
for i in range(len(lats)):
    mlats.append(coord2mlat(lats[i],lons[i],mlat,mlon))
# オーロラが見える最低緯度をKp-indexの1次式で近似
def linear_func(kp, a, b):
    return a*kp+b
# 1次近似係数＆関数を決定
popt, _ = curve_fit(linear_func,kps,mlats)
def kp2mlat(kp):
    return linear_func(kp, popt[0], popt[1])
# オーロラが見える領域を可視化する
fig = plt.figure(figsize=(10,8))
proj = ccrs.PlateCarree(central_longitude=135)
# proj = ccrs.Orthographic(0,90)
ax = fig.add_subplot(1,1,1, projection=proj)
#ax.set_extent([100.0, 170.0, 20.0, 80.0])
ax.set_global(); ax.stock_img()
ax.add_feature(cfeature.COASTLINE, edgecolor='black')
ax.add_feature(cfeature.BORDERS, edgecolor='black')
ax.gridlines(draw_labels=True, crs=proj)
# Kp指数3から9までの、オーロラが見える領域を描く
for kp,c in [[9,'red'],[7,'yellow'],[5,'green'],[3,'blue']]: 
    ax.tissot(rad_km=mlat2km(kp2mlat(kp)),lons=[mlon,],
      lats=[mlat,],fc=c,n_samples=128,ec="black",alpha=0.5)
ax.tissot(rad_km=2000, lons=[mlon,],lats=[mlat,],fc="black",
          n_samples=128,ec="black",alpha=0.5)
# 特定地点から見通せる空を描く場合（京都は北緯35.00、東経135.76）
#ax.tissot(rad_km=1000,lons=[135.76,],lats=[35,],
#    fc="red", n_samples=128, ec="black", alpha=0.5)
plt.show()
```

### ステップバイステップで実行する（＆記事用の画像を生成する）ためのコード

オーロラが見える緯度（磁極中心）を、北半球前提のもとに計算する。

```python
from geopy.distance import geodesic # 緯度経度から距離を算出するのに使う
from scipy.optimize import curve_fit

# 2点間の緯度経度から緯度相当値を算出
def coord2mlat(lat,lon,mlat,mlon): # 指定場所の緯度・経度と、北磁極の緯度・経度
    return 90-geodesic((lat,lon),(mlat,mlon)).kilometers/10000*90
# 緯度相当値mlatから距離を算出
def mlat2km(mlat):
    return 10000.0/90*(90-mlat)
# Kp-indexが5でSeattleやTorontoで見える。7でDenverやNashville、9でTexasやNorthern Mexico
# 磁気緯度65度と70度の間がオーロラ帯
kps  = np.array([5,       5,      7,       7,      9])
lons = np.array([-122.30, -79.40, -104.98, -86.78, -97.03])
lats = np.array([47.45,   43.65,  39.76,   36.16,  32.90])
mlats=[]                # 指定場所と北磁極間の角度を格納する
mlat=81.3; mlon=-110.8  # 現在の北磁極は北緯81.3度 西経110.8度
for i in range(len(lats)):    
    mlats.append(coord2mlat(lats[i],lons[i],mlat,mlon))
# オーロラが見える最低緯度min_latをKp-indexの1次式で近似
def linear_func(kp, a, b):
    min_lat=a*kp+b
    return min_lat
# 1次近似係数＆関数を決定
popt, _ = curve_fit(linear_func,kps,mlats)
def kp2mlat(kp):
    return linear_func(kp, popt[0], popt[1])

for kp in [3,5,7,9]:
    print('Kp=',kp,'なら、磁極緯度が65度〜',int(kp2mlat(kp)),
      '度、北磁極からの距離',int(mlat2km(kp2mlat(kp))),'kmまでオーロラを見ることができる。')
```

## 上空のオーロラを見通せる距離の概算

```python
import math

def distance2horizon(height_km):
    r=63700; 
    return math.sqrt(2*r*height_km+height_km**2)
print('上空600kmを地平線上に見通せる距離(km)',distance2horizon(600))
print('上空600kmを30度見上げて見える距離(km)',600/math.tan(30/180*math.pi))
```

## 1700 年代以降の太陽活動の推移を表示するコード例

```python
import astropy.units as u; import sunpy.timeseries as ts
from sunpy.net import Fido; from sunpy.net import attrs as a; from sunpy.time import TimeRange
from astropy.visualization import time_support

time_range=TimeRange("1749-01-01 00:00",Time.now())
result=Fido.search(a.Time(time_range),a.Instrument('noaa-indices'))
f_noaa_indices=Fido.fetch(result)
noaa=ts.TimeSeries(f_noaa_indices, source='noaaindices').truncate(time_range)
time_support();fig = plt.figure(figsize=(15,2)); ax = plt.axes()
ax.axvspan(datetime(1770+111,7,17),datetime(1770+111,11,17),label="1770/9/17",color="red",alpha=0.9)
ax.axvspan(datetime(1859+111,7,1), datetime(1859+111,11,1),label="1859/9/1(Carrington Event )",
           color="green", alpha=0.9)
ax.plot(noaa.time, noaa.quantity('sunspot RI'), label='Sunspot Number')
ax.set_ylim(bottom=0); ax.set_ylabel('Sunspot Number'); ax.set_xlabel('Year')
ax.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=0)
ax.grid(True); plt.show()
```
