## 簡易BOS法による気流の可視化処理を行う(iOS/iPadOS利用例)

注意事項: 下記のコードは、iOS/iPadOS上でPythonを動かすことを前提としたコードです。
通常のJupyter/OpenCV環境で動かす際には、

- OpenCVで動画取り込みを行う
- 撮影画像に対して、最初の撮影画像との差分をとる
- 差分画像を表示する

という処理を行います。そうした例が、下部にあります。

```python
from uikit.ui_uiview import *
from avfoundation import take_movie_extended_editted as tm
import time, cv2

ui_view = create_uiview_full(name = 'sample')

def process_image(npimg):
    global processed_frames
    if len(tm.images)==0:
        if tm.processed_frames > 10:
            tm.images.append(npimg)
    elif len(tm.images)==1:
        # 最初の撮影画像との差分をとる
        tmp_img = cv2.absdiff(npimg, tm.images[0])
        npimg[:,:] = 15*tmp_img[:, :]

tm.video_shooting_setup( ui_view,
    'AVCaptureDeviceTypeBuiltInTelephotoCamera',
    'AVCaptureSessionPreset640x480',
    process_image )
```

### 撮影処理を実行する

```python
tm.images.clear(); tm.video_shooting_start(.3)
time.sleep(20); tm.video_shooting_close(ui_view)
```

## 簡易BOS法による気流の可視化処理を行う(通常Jupyter利用例)

Jupyterで完結するようなコードです。
そのため、速度など多少無理がありますが、参考までに添付いたします。

```python
from io import BytesIO
import os

import IPython
from PIL import Image
import cv2 as cv

# IPython.displayの機能を使い画像を表示する
def show(a, fmt='jpeg'):
    f = BytesIO()
    # メモリ上にJPEG画像ファイルを作り・読み込み表示する
    Image.fromarray(a).save(f, fmt)
    IPython.display.display(
        IPython.display.Image(data=f.getvalue()))
```

```python
images=[]

def process_image(npimg):
    if len(images)==0:
        images.append(npimg)
        return npimg
    else:
        tmp_img = cv.absdiff(npimg, images[0])
        return 15*tmp_img[:,:]

video_id=2 # video camera
for i in range(30):  # 適当に回数設定する
    cap = cv.VideoCapture(video_id)
    ret, frame = cap.read()
    if ret:
        frame = cv.cvtColor(frame, cv.COLOR_BGR2GRAY)
        img=process_image(frame)
        frame = cv.cvtColor(img, cv.COLOR_GRAY2BGR)
        show(frame)
        #show(frame)
        IPython.display.clear_output(wait=True)
```
